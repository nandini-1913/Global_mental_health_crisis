```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/Global_Mental_Health_Crisis_Index_2026.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Missing values
df.isnull().sum()
```

```python
# checking for duplicasy
df.duplicated().sum()
df = df.drop_duplicates()
df
```

```python
# plotting histogram
df.hist(figsize=(16,14))
plt.show()
```

```python
# Correlation matrix
corr = df.corr(numeric_only=True)

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
import seaborn as sns

# Clean column names (just in case)
df.columns = df.columns.str.strip()

# Sort data for better visualization
df_dep = df.sort_values(by='depression_pct', ascending=False)
df_anx = df.sort_values(by='anxiety_pct', ascending=False)

# Set style
sns.set(style="whitegrid")
```

```python
# -------------------------------
# Depression Plot
# -------------------------------
plt.figure(figsize=(16, 6))
sns.barplot(
    x='country',
    y='depression_pct',
    data=df_dep,
)

plt.xticks(rotation=90)
plt.title('Depression Percentage by Country', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Depression (%)')

plt.tight_layout()
plt.show()
```

```python
# -------------------------------
#  Anxiety Plot
# -------------------------------
plt.figure(figsize=(16, 6))
sns.barplot(
    x='country',
    y='anxiety_pct',
    data=df_anx,
)

plt.xticks(rotation=90)
plt.title('Anxiety Percentage by Country', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Anxiety (%)')

plt.tight_layout()
plt.show()
```

```python
# Filteing Data for India
india = df[df['country'] == 'India']

print(india[['depression_pct', 'anxiety_pct']])
```

```python
# comparing india with global avg
global_avg = df[['depression_pct', 'anxiety_pct']].mean()

india_values = india[['depression_pct', 'anxiety_pct']].iloc[0]

print("India Values:\n", india_values)
print("\nGlobal Average:\n", global_avg)
```

```python
# Visual Comparison (India vs Global)

labels = ['Depression', 'Anxiety']
india_vals = india_values.values
global_vals = global_avg.values

x = range(len(labels))

plt.figure(figsize=(6,4))
plt.bar(x, india_vals, width=0.4, label='India', align='center')
plt.bar(x, global_vals, width=0.4, label='Global Avg', align='edge')

plt.xticks(x, labels)
plt.ylabel('Percentage (%)')
plt.title('India vs Global Mental Health Comparison')
plt.legend()

plt.show()
```

```python
df['dep_rank'] = df['depression_pct'].rank(ascending=False)
df['anx_rank'] = df['anxiety_pct'].rank(ascending=False)

df[df['country'] == 'India'][['dep_rank', 'anx_rank']]
```
